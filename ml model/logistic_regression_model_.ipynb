## **handel categorial variables**

```python
categorical = [var for var in df.columns if df[var].dtype=='O']
print('There are {} categorical variables\n'.format(len(categorical)))
print('The categorical variables are :\n\n', categorical)
df[categorical]
```

```python
df = df.drop(['temp'], axis=1)
df
```

# Split the data into features (X) and target variable (y)


```python
X = df.drop('pump', axis=1)
y = df['pump']
```

# **Splitting the data into training and testing sets**

```python
from sklearn.model_selection import train_test_split , cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 44)
X_train.shape, X_test.shape
```

# **Initializing logistic regression model**

```python
from sklearn.linear_model import LogisticRegression
LogicRegression = LogisticRegression(max_iter=1000)
```

# Applying cross-validation using 5-fold cross-validation

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(LogicRegression, X_train, y_train, cv=5, scoring='accuracy')
```

# **train model on training dataet**

```python
LogicRegression.fit(X_train, y_train)
```

# **Predict the results**

```python
y_pred = LogicRegression.predict(X_test)
y_pred
```

# **Confusion matrix and classification report**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix\n\n', cm)
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
print('\nTrue Positives(TP) = ', TP)
print('\nTrue Negatives(TN) = ', TN)
print('\nFalse Positives(FP) = ', FP)
print('\nFalse Negatives(FN) = ', FN)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# Calculate precision, recall, and F1-score
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```
